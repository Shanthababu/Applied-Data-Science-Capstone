```python
from bs4 import BeautifulSoup as bsoup
from urllib.request import urlopen as uReq
import requests
import lxml
import pandas as pd
from pandas import DataFrame
import numpy as np
```

```python
# list of postal codes in Canada where the first letter is M.- FROM Wikipedia (Web Scraping)
my_url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
# Requests to get web page details from Wikipedia page 
r=requests.get(my_url)
```

```python
# Parse the htlm with Soup
page=bsoup(r.text,"html.parser")
#page
rtable=page.table
rtable
results=rtable.find_all('tr')
nrows=len(results)
nrows
results=rtable.find_all('tr')
nrows=len(results)
print("###################################################################################")
print("Total Number of postal codes in Canada where the first letter is M is: ",nrows)
print("###################################################################################")
```

```python
header=results[0].text.split()
print("#####################################")
print("Headres from the extracted table")
print (header)
print("#####################################")
```

```python
#Iteration loop to extract all cells into a dataframe df_Loc_of_Neigh_Totonto as Postcode', 'Borough', 'Neighbourhood'

records =[]
n=1
while n < nrows :
    Postcode=results[n].text.split('\n')[1]
    Borough=results[n].text.split('\n')[2]
    Neighborhood=results[n].text.split('\n')[3]
    records.append((Postcode, Borough,Neighborhood))
    n=n+1
print("##########################################################################################")
print("Extreatced Data from the URL inserting into DataFrame in standard table format is python ")
print("##########################################################################################")
df_Loc_of_Neigh_Totonto=pd.DataFrame(records, columns=['PostalCode', 'Borough', 'Neighbourhood'])
df_Loc_of_Neigh_Totonto.head(5)
```
